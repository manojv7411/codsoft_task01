```python
# Training the model using Linear Regression correlated to the variable TV

x_train,x_test,y_train,y_test=train_test_split(adf[['TV']],adf[['Sales']],test_size=0.3,random_state=0)
```

Successfully the Data is splitted into (x,y) train part: 0.7 and (x,y) test part: 0.3

------------------------------------------------------------------------------------

Linear Regression

```python
model=LinearRegression()
model.fit(x_train,y_train)
```

```python
res = model.predict(x_test)
print(res)
```
