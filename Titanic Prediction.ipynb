```python
# importing Labelencoder from sk.learn library
from sklearn.preprocessing import LabelEncoder

# Assuming ds is my original DataFrame

# Changing the Gender Column String values into 0's and 1's
label_encoder = LabelEncoder()
ds['Sex'] = label_encoder.fit_transform(ds['Sex'])
ds.head(10)
```

```python
# Split your data into training and test sets
X = ds[['Pclass', 'Sex']]
Y = ds['Survived']
```

```python
# importing train_test_split from the library
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=0)

# importing LogisticRegression from the Library 
from sklearn.linear_model import LogisticRegression

# Now fit your model with X_train and Y_train
log=LogisticRegression(random_state=0)
log.fit(X_train, Y_train)
```

```python
# print the X_test data 
pre=print(log.predict(X_test))
```
